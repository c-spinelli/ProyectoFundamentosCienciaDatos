## Analisis univariado

### Histogramas para cada variable cuantitativa

```python
quantitative_vars = ['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores', 
                     'Tutoring_Sessions', 'Physical_Activity', 'Exam_Score']
```

```python
for col in quantitative_vars:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, bins=20, color='blue')
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

### Gráficos de barras para variables cualitativas

```python
categorical_vars = ['Parental_Involvement', 'Motivation_Level', 'Peer_Influence', 
                    'Teacher_Quality', 'Learning_Disabilities', 'Parental_Education_Level']
```

```python
for col in categorical_vars:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, y=col, order=data[col].value_counts().index)
    plt.title(f'Distribución de {col}')
    plt.ylabel(col)
    plt.xlabel('Frecuencia')
    plt.show()
```

#### Conteo de categorías para variables cualitativas

```python
for col in categorical_vars:
    print(f"\n{col}:\n{data[col].value_counts()}")
```

## Análisis Bivariado/Multivariado

### Gráficos de dispersión para relaciones cuantitativas

```python
sns.pairplot(data[quantitative_vars], diag_kind='kde', kind='scatter')
plt.show()
```

### Matriz de correlación

```python
correlation_matrix = data[quantitative_vars].corr()
print(correlation_matrix)
```

### Heatmap de la matriz de correlación

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Mapa de calor de correlaciones")
plt.show()
```

### Pair plot para analizar relaciones múltiples

```python
sns.pairplot(data[['Hours_Studied', 'Attendance', 'Exam_Score', 'Sleep_Hours', 'Previous_Scores']], diag_kind='kde')
plt.show()
```

### Boxplot para variables cualitativas vs. cuantitativas

```python
sns.boxplot(x='Motivation_Level', y='Exam_Score', data=data, palette='Set2', hue='Motivation_Level', legend=False)
plt.title("Impacto del Nivel de Motivación en el Puntaje del Examen")
plt.show()
```

### Gráficos Adicionales

```python
sns.violinplot(x='Motivation_Level', y='Exam_Score', data=data, palette='muted', hue='Motivation_Level', legend=False)
plt.title("Distribución de Puntaje por Nivel de Motivación")
plt.show()
```

## Analisis de problemas en los datos

#### Valores faltantes

```python
missing_data = data.isnull().sum()
print("\nValores nulos por columna:")
print(missing_data)
```

```python
missing_percentage = data.isnull().mean() * 100
print("Porcentaje de valores faltantes por columna:")
print(missing_percentage)
```

```python
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Visualización de Valores Faltantes")
plt.show()
```

#### Outliers


```python
for col in quantitative_vars:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data[col], color='orange')
    plt.title(f'Valores atípicos en {col}')
    plt.show()
```

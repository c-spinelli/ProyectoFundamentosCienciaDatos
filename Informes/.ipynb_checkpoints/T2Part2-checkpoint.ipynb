```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('StudentPerformanceFactors.csv')
```

```python
data_description = data.select_dtypes(include=['number']).describe()
```

```python
print("Descripción estadística de los datos cuantitativos:")
```

```python
print(data_description)
```

```python
data_description_cat = data.select_dtypes(exclude=['number']).describe()
print("\nDescripción estadística de los datos cualitativos:")
print(data_description_cat)
```

```python
sns.boxplot(x='Motivation_Level', y='Exam_Score', data=data)
plt.title('Impacto del Nivel de Motivación en el Puntaje del Examen')
plt.show()
```

### Analisis univariado

```python
for column in data.select_dtypes(include=['number']).columns: 
    plt.figure(figsize=(8,4))
    sns.histplot(data[column], kde=True, bins=20, color='blue')
    plt.title(f'Distribucion de {column}')
    plt.show()
```

```python
 for column in data.select_dtypes(exclude=['number']).columns:
    plt.figure(figsize=(8,4))
    sns.countplot(y=data[column], order=data[column].value_counts().index)
    plt.title(f'Distribucion de {column}')
    plt.show()
```

```python
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=data)
plt.title('Relacion entre horas de estudio y puntaje del examen')
plt.show()
```

```python
sns.boxplot(x='Parental_Involvement', y='Exam_Score', data= data)
plt.title('Impacto del involucramiento parental en el puntaje del examen')
plt.show()
```
